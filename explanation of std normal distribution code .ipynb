```python
def stdNBgraph(dataset):
    #to creat a functio of  standard normal distributin graph from dataset of specific column
    #to convert normal distribution to standard normal distribution 
    import seaborn as sns
    #load seaborn for shown graph of stdNB
    mean=dataset.mean()
    #to calculate mean of specific column ssc_p
    std=dataset.std()
    #to calculate std of ssc_p
    values=[i for i in dataset] 
    #values for specific column of ssc_p  to convert list
    z_score=[((j-mean)/std) for j in values]
    #to calculate z_score by z_score fromula with for loop  [ x value of ssc_p is subtract to mean value of ssc_p and divided into std of ssc_p]
    sns.distplot(z_score,kde=True)
    #to show the graph of standed normal distribution of respected column of dataset by calculated z_score value with sns distplot
    sum(z_score)/len(z_score)
```

```python
stdNBgraph(dataset["ssc_p"])
#call function of std nbgraph of dataset with  respected column name 
```
